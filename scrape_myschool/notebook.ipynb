```python
import pandas as pd
from bs4 import BeautifulSoup
import time
import requests
from datetime import datetime
from concurrent.futures import ThreadPoolExecutor, as_completed

# Function to fetch and parse a single page
def fetch_page(subject, exam_year, exam_type, paper_type, page):
    """
    Fetches the content of a single page and parses it with BeautifulSoup.
    
    Args:
        subject (str): The subject to scrape (e.g., 'mathematics').
        exam_year (int): The year of the exam to scrape.
        exam_type (str): The type of the exam (e.g., 'obj').
        page (int): The page number to fetch.
    
    Returns:
        BeautifulSoup: Parsed HTML content of the page.
        bool: True if the page fetch is successful, False otherwise.
    """
    url = f'https://myschool.ng/classroom/{subject}?exam_type={exam_type}&exam_year={exam_year}&type={paper_type}&topic=&page={page}'
    response = requests.get(url)

    if response.status_code != 200:
        print(f"Failed to fetch page {page}. Status code: {response.status_code}")
        return None, False

    soup = BeautifulSoup(response.content, 'html.parser')
    return soup, True

# Function to extract data from a page
def extract_data(soup):
    """
    Extracts questions, options, solutions, and exam years from the parsed page content.
    
    Args:
        soup (BeautifulSoup): Parsed HTML content of the page.
    
    Returns:
        dict: A dictionary with extracted questions, options, solutions, and exam years.
    """
    # Extract exam years
    exam_year_elements = soup.find_all(class_='ml-2 badge bg-success text-light')
    exam_year_texts = [element.text.strip() for element in exam_year_elements]

    # Extract questions
    question_elements = soup.find_all(class_='question-desc mt-0 mb-3')
    questions_text = [question.get_text(strip=True) for question in question_elements]

    # Extract options
    option_elements = soup.find_all(class_='list-unstyled')
    all_options = []
    for ul in option_elements:
        question_options = {}
        for li in ul.find_all('li'):
            key = li.find('strong').text.strip().rstrip('.')  # Extract option letter (e.g., A, B)
            value = li.text.split('.', 1)[-1].strip()        # Extract the corresponding value
            question_options[key] = value
        all_options.append(question_options)

    # Extract solution URLs
    solution_elements = soup.find_all(class_='btn btn-sm btn-outline-danger')
    links = [link['href'] for link in solution_elements]

    return {
        'questions': questions_text,
        'options': all_options,
        'solution': links,
        'exam_year': exam_year_texts,
    }

# Function to scrape a single page and extract data
def scrape_page(subject, exam_year, exam_type, paper_type, page):
    soup, success = fetch_page(subject, exam_year, exam_type, paper_type, page)
    if success and soup:
        return extract_data(soup)
    return None

# Main scraping function using ThreadPoolExecutor for parallel processing
def scrape_all_pages_parallel(subject, exam_year, exam_type, paper_type, total_pages=20):
    """
    Scrapes all pages of data for a given subject, exam year, and exam type in parallel.
    
    Args:
        subject (str): The subject to scrape.
        exam_year (int): The year of the exam to scrape.
        exam_type (str): The type of the exam (e.g., 'obj').
        paper_type (str): The paper type (e.g., 'obj').
        total_pages (int): The total number of pages to scrape.
    
    Returns:
        pd.DataFrame: A DataFrame containing all scraped data.
    """
    all_data = {
        'questions': [],
        'options': [],
        'solution': [],
        'exam_year': []
    }

    # Using ThreadPoolExecutor to scrape pages in parallel
    with ThreadPoolExecutor(max_workers=10) as executor:
        futures = []
        for page in range(1, total_pages + 1):
            futures.append(executor.submit(scrape_page, subject, exam_year, exam_type, paper_type, page))
        
        # Collect the results as they complete
        for future in as_completed(futures):
            data = future.result()
            if data:
                all_data['questions'].extend(data['questions'])
                all_data['options'].extend(data['options'])
                all_data['solution'].extend(data['solution'])
                all_data['exam_year'].extend(data['exam_year'])

    return pd.DataFrame(all_data)

# Run the scraper and measure execution time
if __name__ == "__main__":
    subject = 'mathematics'
    exam_year = 2023
    exam_type = 'waec'
    paper_type = 'obj'
    total_pages = 20  # You can adjust this number to scrape more or fewer pages

    start_time = datetime.now()  # Start timer

    # Scrape data
    examination_df = scrape_all_pages_parallel(subject, exam_year, exam_type, paper_type, total_pages)

    end_time = datetime.now()  # End timer
    elapsed_time = end_time - start_time

    # Display the DataFrame and runtime
    print(f"Scraping completed in {elapsed_time}.")

    # Save to CSV
    examination_df.to_csv('examination.csv', index=False)
    print("Data saved to 'examination.csv'.")
```

```python
import pandas as pd
from bs4 import BeautifulSoup
import time
import requests
from datetime import datetime
from concurrent.futures import ThreadPoolExecutor, as_completed
import json

# Function to fetch and parse a single page
def fetch_page(subject, exam_year, exam_type, paper_type, page):
    """
    Fetches the content of a single page and parses it with BeautifulSoup.
    
    Args:
        subject (str): The subject to scrape (e.g., 'mathematics').
        exam_year (int): The year of the exam to scrape.
        exam_type (str): The type of the exam (e.g., 'waec').
        paper_type (str): The type of the paper (e.g., 'obj').
        page (int): The page number to fetch.

    Returns:
        BeautifulSoup: Parsed HTML content of the page.
        bool: True if the page fetch is successful, False otherwise.
    """
    url = f'https://myschool.ng/classroom/{subject}?exam_type={exam_type}&exam_year={exam_year}&type={paper_type}&topic=&page={page}'
    response = requests.get(url)

    # Check if the page was fetched successfully
    if response.status_code != 200:
        print(f"Failed to fetch page {page} for year {exam_year}. Status code: {response.status_code}")
        return None, False

    # Parse the page content with BeautifulSoup
    soup = BeautifulSoup(response.content, 'html.parser')
    return soup, True

# Function to extract data from a page
def extract_data(soup):
    """
    Extracts questions, options, solutions, and exam years from the parsed page content.
    
    Args:
        soup (BeautifulSoup): Parsed HTML content of the page.
    
    Returns:
        dict: A dictionary containing extracted questions, options, solutions, and exam years.
    """
    # Extract exam years from the page
    exam_year_elements = soup.find_all(class_='ml-2 badge bg-success text-light')
    exam_year_texts = [element.text.strip() for element in exam_year_elements]

    # Extract questions from the page
    question_elements = soup.find_all(class_='question-desc mt-0 mb-3')
    questions_text = [question.get_text(strip=True) for question in question_elements]

    # Extract options for each question
    option_elements = soup.find_all(class_='list-unstyled')
    all_options = []
    for ul in option_elements:
        question_options = {}
        for li in ul.find_all('li'):
            key = li.find('strong').text.strip().rstrip('.')  # Option letter (e.g., A, B)
            value = li.text.split('.', 1)[-1].strip()        # Option value (e.g., Option A: Answer)
            question_options[key] = value
        all_options.append(question_options)

    # Extract solution URLs for the questions
    solution_elements = soup.find_all(class_='btn btn-sm btn-outline-danger')
    links = [link['href'] for link in solution_elements]

    # Return all extracted data in a dictionary
    return {
        'questions': questions_text,
        'options': all_options,
        'solution': links,
        'exam_year': exam_year_texts,
    }

# Function to scrape a single page and extract data
def scrape_page(subject, exam_year, exam_type, paper_type, page):
    """
    Scrapes a single page and extracts the data.
    
    Args:
        subject (str): The subject to scrape.
        exam_year (int): The year of the exam to scrape.
        exam_type (str): The type of the exam (e.g., 'waec').
        paper_type (str): The type of the paper (e.g., 'obj').
        page (int): The page number to scrape.

    Returns:
        dict: Extracted data from the page, or None if the page fetch failed.
    """
    soup, success = fetch_page(subject, exam_year, exam_type, paper_type, page)
    if success and soup:
        return extract_data(soup)
    return None

# Main scraping function using ThreadPoolExecutor for parallel processing
def scrape_all_pages_parallel(subject, exam_year, exam_type, paper_type, total_pages=20):
    """
    Scrapes all pages of data for a given subject, exam year, and exam type in parallel.
    
    Args:
        subject (str): The subject to scrape.
        exam_year (int): The year of the exam to scrape.
        exam_type (str): The type of the exam (e.g., 'waec').
        paper_type (str): The type of the paper (e.g., 'obj').
        total_pages (int): The total number of pages to scrape.
    
    Returns:
        pd.DataFrame: A DataFrame containing all the scraped data.
    """
    # Initialize an empty dictionary to store the scraped data
    all_data = {
        'questions': [],
        'options': [],
        'solution': [],
        'exam_year': []
    }

    # Using ThreadPoolExecutor to scrape pages in parallel
    with ThreadPoolExecutor(max_workers=10) as executor:
        futures = []
        # Submit each page to be scraped in parallel
        for page in range(1, total_pages + 1):
            futures.append(executor.submit(scrape_page, subject, exam_year, exam_type, paper_type, page))
        
        # Collect the results as they complete
        for future in as_completed(futures):
            data = future.result()
            if data:
                all_data['questions'].extend(data['questions'])
                all_data['options'].extend(data['options'])
                all_data['solution'].extend(data['solution'])
                all_data['exam_year'].extend(data['exam_year'])

    # Convert the collected data to a DataFrame
    return pd.DataFrame(all_data)

# Loop through the years from start_year to end_year and scrape data for each year
def scrape_for_years(subject, start_year, end_year, exam_type, paper_type, total_pages=20):
    """
    Scrapes data for multiple years in the specified range.
    
    Args:
        subject (str): The subject to scrape.
        start_year (int): The starting year for scraping.
        end_year (int): The ending year for scraping.
        exam_type (str): The type of the exam (e.g., 'waec').
        paper_type (str): The type of the paper (e.g., 'obj').
        total_pages (int): The total number of pages to scrape for each year.
    
    Returns:
        pd.DataFrame: A DataFrame containing all scraped data for the years in the range.
    """
    all_exam_data = pd.DataFrame()

    # Loop through each year in the specified range
    for exam_year in range(start_year, end_year + 1):
        print(f"Scraping data for year: {exam_year}...")
        # Scrape the data for the current year
        examination_df = scrape_all_pages_parallel(subject, exam_year, exam_type, paper_type, total_pages)
        # Append the data to the combined DataFrame
        all_exam_data = pd.concat([all_exam_data, examination_df], ignore_index=True)

    return all_exam_data

# Function to save the DataFrame to a prettified JSON file (row by row)
def save_to_pretty_json_row_by_row(df, filename):
    """
    Saves each row of the DataFrame as a prettified JSON object to a file.
    
    Args:
        df (pd.DataFrame): The DataFrame containing the data to be saved.
        filename (str): The name of the file to save the JSON data.
    """
    with open(filename, 'w', encoding='utf-8') as f:
        # Loop through each row of the DataFrame
        for _, row in df.iterrows():
            # Convert each row to a dictionary
            row_dict = row.to_dict()

            # Prettify and write each row as a JSON object with indentation, no escaping of special characters
            json.dump(row_dict, f, indent=4, ensure_ascii=False)  # Set ensure_ascii to False to avoid unicode escapes
            f.write('\n')  # Add a new line after each JSON object

# Run the scraper and measure execution time
if __name__ == "__main__":
    subject = 'mathematics'
    start_year = 1988
    end_year = 2023
    exam_type = 'waec'
    paper_type = 'obj'
    total_pages = 20  # You can adjust this number to scrape more or fewer pages

    # Start the timer to measure the execution time
    start_time = datetime.now()  # Start timer

    # Scrape data for all years in the specified range
    all_examination_df = scrape_for_years(subject, start_year, end_year, exam_type, paper_type, total_pages)

    # End the timer and calculate the elapsed time
    end_time = datetime.now()  # End timer
    elapsed_time = end_time - start_time

    # Display the DataFrame and runtime
    print(f"Scraping completed in {elapsed_time}.")

    # Save the scraped data to a prettified JSON file (row by row)
    save_to_pretty_json_row_by_row(all_examination_df, 'all_examination_data.json')
    print("Data saved to 'all_examination_data.json'.")
```

```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
from datetime import datetime
from concurrent.futures import ThreadPoolExecutor, as_completed
import json
import re
from urllib.parse import urljoin


def fetch_question_page(url):
    """
    Fetches and parses the page content for a given URL.
    """
    try:
        headers = {
            'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
        }
        response = requests.get(url, headers=headers, timeout=10)
        if response.status_code != 200:
            print(f"Failed to fetch URL: {url}. Status code: {response.status_code}")
            return None, False
        soup = BeautifulSoup(response.content, 'html.parser')
        return soup, True
    except requests.RequestException as e:
        print(f"Error fetching URL {url}: {e}")
        return None, False


def extract_question_data(id, soup, base_url):
    """
    Extracts question data from the given BeautifulSoup object.
    """
    if not soup:
        return {
            'id': id,
            'subject_year': None,
            'topic_year': None,
            'question': None,
            'question_diagram': None,
            'options': [],
            'correct_answer': None,
            'explanation': None,
            'url': base_url
        }

    # Extract question text
    question_text = None
    parent_div = soup.find('div', class_='question-desc mb-3')
    if parent_div:
        p_tags = parent_div.find_all('p')
        question_text = ' '.join([p.get_text(strip=True) for p in p_tags]) if p_tags else None

    # Extract question image URL
    img_tag = parent_div.find('img', class_='img-fluid') if parent_div else None
    img_url = urljoin(base_url, img_tag['src']) if img_tag and 'src' in img_tag.attrs else None

    # Extract correct answer
    correct_answer_tag = soup.find_all(class_="text-success mb-3")
    correct_answer = [tag.text.strip('Option')[0] for tag in correct_answer_tag] if correct_answer_tag else None

    # Extract options
    options_container = soup.find('ul', class_='list-unstyled')
    options_dict = {}
    if options_container:
        options = options_container.find_all('li')
        for option in options:
            key = option.find('strong').text.strip('.').strip() if option.find('strong') else None
            value = option.text.split(".")[-1].strip() if option.text else None
            if key and value:
                options_dict[key] = value

    # Extract explanation
    explanation_texts = []
    explanation_tags = soup.find_all(class_="mb-4")
    for tag in explanation_tags:
        text = tag.get_text(strip=True)
        if "Contributions" in text:
            break
        explanation_texts.append(text)
    explanation_content = ' '.join(explanation_texts).split("Explanation", 1)[-1].strip() if "Explanation" in ' '.join(explanation_texts) else None

    # Extract subject and year from URL
    match = re.search(r'/classroom/([^/]+).*exam_year=(\d+).*type=([^&]+)', base_url)
    subject_year = f"{match.group(1).capitalize()} {match.group(3).capitalize()} {match.group(2)}" if match else None

    # Return all extracted data in a dictionary
    return {
        "id": id,
        "subject_year": subject_year,
        "topic_year": "",
        "question": question_text,
        "question_diagram": img_url,
        "options": options_dict,
        "correct_answer": correct_answer,
        "explanation": f"Hint: {explanation_content}" if explanation_content else None,
        "url": base_url
    }


def scrape_urls_in_batches(id_list, urls, batch_size=20):
    """
    Scrapes the given URLs in batches and returns a DataFrame of results.
    """
    results = []

    with ThreadPoolExecutor(max_workers=batch_size) as executor:
        future_to_url = {executor.submit(fetch_question_page, url): (id_list[i], url) for i, url in enumerate(urls)}

        for future in as_completed(future_to_url):
            id_, url = future_to_url[future]
            try:
                soup, success = future.result()
                data = extract_question_data(id_, soup, base_url=url) if success else {
                    'id': id_,
                    'subject_year': None,
                    'topic_year': None,
                    'question': None,
                    'question_diagram': None,
                    'options': [],
                    'correct_answer': None,
                    'explanation': None,
                    'url': url
                }
                results.append(data)
            except Exception as e:
                print(f"Error processing URL {url}: {e}")
                results.append({
                    'id': id_,
                    'subject_year': None,
                    'topic_year': None,
                    'question': None,
                    'question_diagram': None,
                    'options': [],
                    'correct_answer': None,
                    'explanation': None,
                    'url': url
                })

    return pd.DataFrame(results)


def save_to_pretty_json_row_by_row(df, filename):
    """
    Saves each row of the DataFrame as a prettified JSON object to a file
    encoded for LaTeX and Unicode compatibility.
    """
    with open(filename, 'w', encoding='utf-8') as f:
        for _, row in df.iterrows():
            json.dump(row.to_dict(), f, indent=4, ensure_ascii=False)
            f.write('\n')


# Main execution
if __name__ == "__main__":
    csv_file = 'mathematics_waec_obj_1988-2023_test.csv'  # Path to the CSV containing URLs
    batch_size = 20  # Adjust batch size between 10-20 as needed

    # Load URLs from the CSV file
    question_url_df = pd.read_csv(csv_file)
    urls = question_url_df['question_url'].tolist()
    ids = question_url_df['id'].tolist()

    start_time = datetime.now()

    # Scrape all data
    all_examination_df = scrape_urls_in_batches(ids, urls, batch_size=batch_size)

    end_time = datetime.now()
    elapsed_time = end_time - start_time
    print(f"Scraping completed in {elapsed_time}.")
    print(f"Total questions scraped: {len(all_examination_df)}")

    # Save data to JSON
    save_to_pretty_json_row_by_row(all_examination_df, 'scraped_data.json')
    print("Data saved to 'scraped_data.json'.")
```
