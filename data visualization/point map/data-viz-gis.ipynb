```python
# import all the required library
#import geopandas as an alias
import geopandas as gpd  #gpd is the alias

 # we did not import the whole folium libary, instead we import functions we needed from the libary
from folium import Map
from folium import Marker
from folium.plugins import HeatMap, MarkerCluster
```

```python
# read the data using geopandas function -> gpd.read_file
# provide the location to your data inside the function
crime = gpd.read_file(r"D:\New Task\1_Assessment online gis brief\London  Crime Data 2022.csv")

# the head function helps us to have a snippet of our data by showing us the five top rows
crime.head()
```

```python
# Select the Burglary data and visualize it 
Burglary_data = crime[crime["Crime type"] == "Burglary"]
Burglary_data.head()
```

```python
# the first folium function is to call out the basemap
map_1 = Map(location=[51.509969, -0.070510], tiles='openstreetmap', zoom_start=14)

# Visualize the cime data
for idx, row in Burglary_data.iterrows():
    Marker([row['Latitude'], row['Longitude']]).add_to(map_1)
    
map_1 
```

```python
# Create the map
map_2 = Map(location=[51.509969, -0.070510], tiles='openstreetmap', zoom_start=13)

# Add points to the map
mc = MarkerCluster()
for idx, row in Burglary_data.iterrows():
    mc.add_child(Marker([row['Latitude'], row['Longitude']]))
map_2.add_child(mc)

# Display the map
map_2
```

```python
# Create a base map
map_3 = Map(location=[51.509969, -0.070510], tiles='openstreetmap', zoom_start=12)

# Add a heatmap to the base map
HeatMap(data=Burglary_data[['Latitude', 'Longitude']], radius=10).add_to(map_3)

map_3
```
